```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("p24_selected_variables_kor.csv", encoding='utf-8-sig')
df = df.replace(' ', pd.NA)

### 이진형 매핑들 (성별, 직업, 스마트폰보유, AI이용여부)

# 성별 (남:0, 여:1)
df['성별'] = df['성별'].map({1: 0, 2: 1})

# 직업 (있음:1, 없음:0)
df['직업'] = df['직업'].map({1: 1, 2: 0})

# 스마트폰 보유
df['스마트폰보유'] = df['스마트폰보유'].map({1: 1, 2: 0})

# AI 이용여부
df['AI이용여부'] = df['AI이용여부'].map({1: 1, 2: 0})


### 소득 그룹화 (저:0, 중:1, 고:2)
df['소득그룹'] = df['소득'].apply(lambda x: 0 if x <= 3 else (1 if x <= 6 else 2))

### OTT이용빈도 매핑 (8단계 → 8점 척도)
ott_mapping = {
    1: 8, 2: 7, 3: 6, 4: 5, 5: 4, 6: 3, 7: 2, 8: 1
}
df['OTT이용빈도'] = df['OTT이용빈도'].map(ott_mapping)

### SNS좋아요댓글 매핑 (12단계 → 적극성 척도)
sns_mapping = {
    1: 10, 2: 9, 3: 8, 4: 7, 5: 6, 6: 5, 7: 4, 8: 3, 
    9: 2, 10: 12, 11: 11, 12: 10
}
df['SNS좋아요댓글'] = df['SNS좋아요댓글'].map(sns_mapping)

sns_feed_mapping = {
    9: 12,  # 하루 5회 이상
    10: 11, # 하루 3~4회
    11: 10, # 하루 1~2회
    1: 9,   # 하루에도 여러 번
    2: 8,
    3: 7,
    4: 6,
    5: 5,
    6: 4,
    7: 3,
    8: 2
}
df['SNS피드확인'] = df['SNS피드확인'].map(sns_feed_mapping)


# 리커트형 (검색능력, 정보판별1~4, AI인지수준)은 그대로 사용 (1~5 또는 1~4 점수)

# 결과 확인
print(df.head())

# 전처리 완료 저장 (선택)
df.to_csv("p24_fully_mapped.csv", index=False, encoding='utf-8-sig')
```
